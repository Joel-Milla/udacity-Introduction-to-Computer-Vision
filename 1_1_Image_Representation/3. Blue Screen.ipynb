# Color Threshold, Blue Screen

### Import resources

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

### Read in and display the image

```python
image = cv2.imread('./images/pizza_bluescreen.jpg')

# print info image
image.shape
```

```python
# print type of the image
type(image)
```

```python
image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

### Define the color threshold

```python
# define lower and upper bound for color that want to isolate. to select the pixels that have this color and delete it
lower_blue = np.array([[0, 0, 230]]) # define 
upper_blue = np.array([[50, 50, 255]]) #its okay to have nor red or green
```

### Create a mask

```python
mask = cv2.inRange(image_copy, lowerb=lower_blue, upperb=upper_blue) # if the pixel doesn't fall into this range, will block it out

plt.imshow(mask, cmap='gray')

# The white is the part of the image that will be allowed to change, meaning that will be allowed to show through
# the black part is the part of the image that will be blocked, meaning it will stay static with the same image. 
```
